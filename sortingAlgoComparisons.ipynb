# Sort Algorithms Runtime Comparision

## Imports

```python
import math
import time
import random as rand
import matplotlib.pyplot as plt
from copy import deepcopy
```

Import the indivudual sorts. The individual python files are located in the git repo

```python
from heap_sort import HEAPSORT
from insertion_sort import INSERTION_SORT
from merge_sort import MERGE_SORT
from quick_sort import QUICKSORT
```

## Set Up

```python
sorts = {
    "heapsort":[],
    "insertion":[],
    "merge":[],
    "quick":[]
}
```

```python
# setting up the experiment

# the sizes of each dataset
array_sizes = [2**x for x in range(0, 10)]

# set the seed
rand.seed(42)
```

## Running all the sorting algorithms

```python
for size in array_sizes:
    sort_base = [0] * size;
    
    for ii in range(size):
        sort_base[ii] = rand.randint(0, 256)
        
    print("heap sort with %d elements" % size)
    sort = deepcopy(sort_base)
    start = time.time()
    HEAPSORT(sort)
    end = time.time()
    print("Time Elapsed: ", end-start)
    sorts["heapsort"].append(end-start)    
    
    print("insertion sort with %d elements" % size)
    sort = deepcopy(sort_base)
    start = time.time()
    INSERTION_SORT(sort)
    end = time.time()
    print("Time Elapsed: ", end-start)
    sorts["insertion"].append(end-start)
    
    print("Merge sort with %d elements" % size)
    sort = deepcopy(sort_base)
    start = time.time()
    MERGE_SORT(sort, 0, len(sort)-1)
    end = time.time()
    print("Time Elapsed: ", end-start)
    sorts["merge"].append(end-start)
    
    print("Quick sort with %d elements" % size)
    sort = deepcopy(sort_base)
    start = time.time()
    QUICKSORT(sort, 0, len(sort)-1)
    end = time.time()
    print("Time Elapsed: ", end-start)
    sorts["quick"].append(end-start)
    
    
```

## Plotting the Runtimes for each algorithm

```python
plt.figure(figsize=(18, 15))

for key in sorts.keys():
    plt.plot(array_sizes, sorts[key], label=key)
    
plt.legend()
plt.xlabel("Array Sizes")
plt.ylabel("Run Times")
plt.title("Runtime Comparison")
plt.autoscale(enable=True, axis='x')
# plt.xticks([x for x in range(0, max(array_sizes)+10, 10)])

plt.grid()
plt.show()
```
